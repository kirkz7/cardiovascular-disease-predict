```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
from package import function
%matplotlib inline
```

```python
Xtrain = pd.read_csv("Xtrain.csv")
ytrain = pd.read_csv("ytrain.csv").cardio
Xtest = pd.read_csv("Xtest.csv")
ytest = pd.read_csv("ytest.csv").cardio
```

```python
neigh_baseline = KNeighborsClassifier(n_neighbors = 1,p=1,weights = 'distance')
neigh_baseline.fit(Xtrain,ytrain)
ypr = neigh_baseline.predict(Xtest)
probTest = neigh_baseline.predict_proba(Xtest)
probTest = probTest[:,1]
function.eval_confusion("KNN baseline",neigh_baseline.score(Xtest,ytest),ypr,ytest)
```

```python
function.eval_auc("KNN baseline",ytest,probTest)
```

```python
neigh_baseline
```

```python
error_rate = []
for i in range(1,50):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(Xtrain,ytrain)
    pred_i = knn.predict(Xtest)
    error_rate.append(np.mean(pred_i != ytest))

plt.figure(figsize=(10,6))
plt.plot(range(1,50),error_rate)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
print("Minimum error:",min(error_rate),"at K =",error_rate.index(min(error_rate)))
```

```python
neigh = KNeighborsClassifier()
param_grid={'n_neighbors':[30,50],'weights': ('uniform', 'distance'),'p':(1,2)}
clf = GridSearchCV(neigh, param_grid)
clf.fit(Xtrain,ytrain)
clf.best_params_
```

```python
plt.figure(figsize=(10,6))
for i in range(40,51):
    neigh = KNeighborsClassifier(n_neighbors = i,p=1,weights = 'distance')
    cvs = cross_val_score(neigh,Xtrain,ytrain,cv=5)
    plt.plot(cvs,label = "n_neighbors = " + str(i))
plt.legend()
plt.show()
```

```python
neigh_final = KNeighborsClassifier(n_neighbors = 49,p=1,weights = 'distance')
neigh_final.fit(Xtrain,ytrain)
probTest = neigh_final.predict_proba(Xtest)
probTest = probTest[:,1]
function.eval_confusion("KNN p=1",neigh_final.score(Xtest,ytest),ypr,ytest)
```

```python
function.eval_auc("KNN p=1",ytest,probTest)
```
