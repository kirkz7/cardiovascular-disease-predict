```python
import pandas as pd
import numpy as np
from package import function
from package import DT
```

```python
Xtrain = pd.read_csv("Xtrain.csv")
ytrain = pd.read_csv("ytrain.csv").cardio
Xtest = pd.read_csv("Xtest.csv")
ytest = pd.read_csv("ytest.csv").cardio
```

```python
Xtrain["cardio"] = ytrain
Xtrain_male = Xtrain.query("gender == 2").drop(["cardio","gender"],axis = 1)
ytrain_male = Xtrain.query("gender == 2").cardio
Xtest["cardio"] = ytest
Xtest_male = Xtest.query("gender == 2").drop(["cardio","gender"],axis = 1)
ytest_male = Xtest.query("gender == 2").cardio
```

```python
Xtrain_female = Xtrain.query("gender == 1").drop(["cardio","gender"],axis = 1)
ytrain_female = Xtrain.query("gender == 1").cardio
Xtest_female = Xtest.query("gender == 1").drop(["cardio","gender"],axis = 1)
ytest_female = Xtest.query("gender == 1").cardio
```

```python
model = DT.DecisionTree()
model.baseline()
model.best_model(Xtrain_male,ytrain_male)
model.fit(Xtrain_male,ytrain_male)
ypr,probTest = model.predict(Xtest_male)
function.eval_confusion("DTF",ypr,ytest_male)
```

```python
model = DT.DecisionTree()
model.baseline()
model.best_model(Xtrain_female,ytrain_female)
model.fit(Xtrain_male,ytrain_male)
ypr,probTest = model.predict(Xtest_female)
function.eval_confusion("DTF",ypr,ytest_female)
```
